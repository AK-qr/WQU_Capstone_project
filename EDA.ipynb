### Trade sizes statistics

Now let's calculate statistics such as 0.05 and 0.95 quantiles, median, mean, variance for trade sizes.

```python
def get_statistics(array):
    # Quantile 0.05
    q_005 = np.quantile(array, 0.05)

    # Quantile 0.95
    q_095 = np.quantile(array, 0.95)

    # Median
    med = array.median()

    # Mean
    mean = array.mean()

    # Dispersion
    disp = array.std()

    return {
        'quantile_0.05': q_005,
        'quantile_0.95': q_095,
        'median': med,
        'mean': mean,
        'dispersion': disp
    }
```

```python
print('BTC-size\n', get_statistics(trades_btc['size']), end='\n\n')
print('BTC-size(USD)\n', get_statistics(trades_btc['size']*trades_btc['price']), end='\n\n')

print()

print('ETH-size\n', get_statistics(trades_eth['size']), end='\n\n')
print('ETH-size(USD)\n', get_statistics(trades_eth['size']*trades_eth['price']), end='\n\n')
```

Bitcoin has about 2 times higher values (than ether) relative to (USD). Probably high popularity attracts bigger players.

### Order book "penetration"

Also, we need to plot a graph of the order book’s "penetration" magnitude versus the trade size—that is, how much the order book is penetrated as a result of a trade. The penetration magnitude is defined as the change in the best price resulting from the trade. This should be calculated separately for each side.

By order book penetration, we mean "retrospective penetration." That is, how and when the order book was penetrated during trading, rather than how much it can be penetrated at a given moment (in which case the graph would coincide with the ask/bid_cumvol_ graph).

```python
for i in range(10):
    d = dict(zip(lobs_btc[' exchange_ts'], lobs_btc['ask_cumvol_' + str(i)]))
    trades_btc['ask_cumvol_' + str(i)] = trades_btc['exchange_ts'].map(d)

    d = dict(zip(lobs_btc[' exchange_ts'], lobs_btc['bid_cumvol_' + str(i)]))
    trades_btc['bid_cumvol_' + str(i)] = trades_btc['exchange_ts'].map(d)

    d = dict(zip(lobs_btc[' exchange_ts'], lobs_btc['btcusdt:Binance:LinearPerpetual_ask_price_' + str(i)]))
    trades_btc['ask_price_' + str(i)] = trades_btc['exchange_ts'].map(d)

    d = dict(zip(lobs_btc[' exchange_ts'], lobs_btc['btcusdt:Binance:LinearPerpetual_bid_price_' + str(i)]))
    trades_btc['bid_price_' + str(i)] = trades_btc['exchange_ts'].map(d)
```

```python
trades_btc_rr = trades_btc.dropna()
trades_btc_rr
```

```python
shift = []
shift_usd = []

# Let's calculate the extent to which the order book was penetrated and how much people overpaid.

for index, row in trades_btc_rr.iterrows():
    ans = 0
    ans_usd = 0
    if row['aggro_side'] == 'ASK':
        for i in range(10):
            if row['ask_cumvol_' + str(i)] < row['size']:
                ans += 1
        ans_usd = row['ask_price_' + str(min(ans, 4))] - row['ask_price_0']
        shift.append(ans)
        shift_usd.append(ans_usd)
    else:
        for i in range(10):
            if row['ask_cumvol_' + str(i)] < row['size']:
                ans += 1
        ans_usd = row['ask_price_' + str(min(ans, 4))] - row['ask_price_0']
        shift.append(-ans)
        shift_usd.append(-ans_usd)

trades_btc_rr['shift'] = shift
trades_btc_rr['shift_usd'] = shift_usd
```
